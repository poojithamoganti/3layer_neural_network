```python
import numpy as np
from matplotlib import pyplot as plt
import random
import time
import math
%matplotlib inline
from IPython import display

# Sigmoid Function
def sigmoid(u):
    return 1 / (1 + np.exp(-u))

# Derivative of sigmoid
def sigmoid_p(u):
    return u*(1-u) 


x1_1_list =[]
x2_1_list =[]
x3_1_list =[]
error_list =[]

x_list = [[0 for x in range (3)] for y in range(100)]
t = []
le =0.4


#creating 50 samples for first data set
for i in range(50):
  x1_1 = random.uniform(-0.5,0.5)
  x2_1 = random.uniform(-0.5,0.5)
  x3_1= 1
  x1_1_list.append(x1_1)
  x2_1_list.append(x2_1)
  x3_1_list.append(x3_1)
  #print(x1_1_list)
  #print(x2_1_list)
  #print(x3_1_list)
  #print( )



  t.append(0)

#creating 50 samples for second data set
#loop to create 25 negative datasets
for i in range(25):
  r = random.uniform(9,16)
  x1_2 = random.uniform(-3,3)
  x2_2 = - math.sqrt(r - (x1_2)**2)
  x3_2 = 1
  x1_1_list.append(x1_2)
  x2_1_list.append(x2_2)
  x3_1_list.append(x3_2)
  t.append(1)
 
 #loop to create 25 positive datasets
for i in range(25):
  r = random.uniform(9,16)
  x1_2 = random.uniform(-3,3)
  x2_2 = math.sqrt(r - (x1_2)**2)
  x3_2 = 1
  x1_1_list.append(x1_2)
  x2_1_list.append(x2_2)
  x3_1_list.append(x3_2)
  t.append(1)
 #adding all x1 ,x2 ,x3 values to a single list  
  for i in range(len(x1_1_list)):
    x_list[i][0] = x1_1_list[i]
    x_list[i][1] = x2_1_list[i]
    x_list[i][2] = x3_1_list[i]
 
print ("x",x_list)




#plotting initial dataset
fig = plt.figure(figsize= (14,14))
ax = fig.add_subplot(1,3,1,projection = '3d')
plt.title('Initial Dataset')
for i in range(len(t)):
  if(t[i]==0):
    ax.scatter( x1_1_list[i] ,x2_1_list[i] ,x3_1_list[i] ,marker = "o" , color = 'blue')
  else:
    ax.scatter(x1_1_list[i] ,x2_1_list[i] ,x3_1_list[i], marker = "s" ,color = 'green')

plt.show()

   

# random values for hidden 1, 2 and output layers
whiddenonelayer = np.random.uniform(-2,2,[3,3])
whiddentwolayer = np.random.uniform(-2,2,[3,4])
woutputlayer = np.random.uniform(-2,2,[4])

# bias values for hidden 1, 2 and output layers
bhiddenonelayer = np.random.uniform(-2.0 ,2, 3)
bhiddentwolayer = np.random.uniform(-2 ,2, 4)
boutputlayer = np.random.uniform(-2.0 ,2, 1)

while True:
  total_error =0

  for i in range(100):
    # first hidden layer
    zlayer = (np.dot(x[i] , whiddenonelayer)) + bhiddenonelayer
    zAct = sigmoid(zlayer)
    # 1 x 3 matrix
    #print(zlayer)
    #print()

    #second hidden layer 
    ulayer = np.dot(zAct.T, whiddentwolayer ) + bhiddentwolayer
    uAct = sigmoid(ulayer)
    # 1x 4 matrix
    #print(ulayer)  
    #print()    

    #output layer
    olayer = np.dot(uAct ,woutputlayer ) + boutputlayer
    oAct = sigmoid(olayer) 
    # 1 x 1 matrix
    #print(oAct)
    #print( )

    # Error function
    individualerror = ( t[i] - oAct)**2
    # 1 x 1 matrix
    #print(individualerror) 


    #total error calculation
    total_error = total_error + individualerror
    

    #derivative of error function
    der_ind_error = (t[i] - oAct)
    # 1 x 1 matrix
    #print(der_ind_error) 

    #backpropagation
    b_woutputlayer = np.multiply((le * der_ind_error * sigmoid_p(oAct)) , uAct)
    # 1 x 4 matrix 
    #print(b_woutputlayer) 

    b_boutputlayer = le * der_ind_error * sigmoid_p(oAct)
    # 1 x 1 matrix 
    #print(b_boutputlayer)         

    # changed weights for output layer
    woutputlayer = woutputlayer + b_woutputlayer
    # 4 x 4 matrix 
    #print(woutputlayer)

    boutputlayer = boutputlayer + b_boutputlayer
    # 1 x 1 matrix 
    #print(boutputlayer)

    # sum
    sum = np.sum(individualerror * sigmoid_p(oAct))
    # # 1 x 1 matrix 
    #print(sum)

    # changed weights for hidden layer 2
    # transpose
    b_whiddentwolayer = np.transpose([le * der_ind_error * sum * sigmoid_p(uAct) * hidden for hidden in zAct ])
    #  transpose : 3 x 4 matrix 
    #print(b_whiddentwolayer)

    b_bhiddentwolayer = sum * sigmoid_p(uAct)
    # 1 x 4 matrix 
    #print(b_bhiddentwolayer)

    #weights change
    whiddentwolayer = whiddentwolayer+ b_whiddentwolayer.T
    # 3 x 4 matrix 
    #print(whiddentwolayer)

    bhiddentwolayer = bhiddentwolayer + b_bhiddentwolayer
    # # 1 x 4 matrix 
    #print(bhiddentwolayer)

    # changed weights for hidden layer 1
    sum_1 = np.sum(sum * sigmoid_p(uAct))
    #print(sum_1)

    # transpose
    b_whiddenonelayer = np.transpose([le * der_ind_error * sum_1 * sigmoid_p(zAct) * input for input in x[i]])
    #  transpose : 3 x 4 matrix 
    #print(b_whiddentwolayer)

    b_bhiddenonelayer = le * sum_1 * sigmoid_p(zAct) * der_ind_error
    # 1 x 4 matrix 
    #print(b_bhiddentwolayer)

    #weights change
    whiddenonelayer = whiddenonelayer + b_whiddenonelayer
    # # 1 x 1 matrix 
    #print(whiddenonelayer)

    bhiddenonelayer = bhiddenonelayer + b_bhiddenonelayer
    # 1 x 1 matrix 
    #print(bhiddenonelayer)

#appending total eror values into a list
  error_list.append(total_error)
  print(total_error)  

  if(total_error<0.01):
    break 


#plotting error graph
plt.plot(range(len(error_list)) , error_list)
plt.title(" Final Error Graph")
plt.ylabel("Error List")
plt.xlabel("Epochs")




 
```
